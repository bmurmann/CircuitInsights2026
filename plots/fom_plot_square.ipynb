```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vov = np.linspace(0.001, 0.5, num=100)
gm_id = 2/vov
gm_cgg = 100*vov
gm_gds = 5/vov

plt.plot(vov, gm_id, label='gm/ID', color='blue', linestyle='-')

plt.plot(vov, gm_cgg, label='gm/cgg', color='red', linestyle='-')
plt.plot(vov, 0.95*gm_cgg, label='gm/cgg', color='red', linestyle='-')
plt.plot(vov, 0.90*gm_cgg, label='gm/cgg', color='red', linestyle='-')
plt.plot(vov, 0.85*gm_cgg, label='gm/cgg', color='red', linestyle='-')

plt.plot(vov, gm_gds, label='gm/gds', color='green', linestyle='-')
plt.plot(vov, 0.95*gm_gds, label='gm/gds', color='green', linestyle='-')
plt.plot(vov, 0.90*gm_gds, label='gm/gds', color='green', linestyle='-')
plt.plot(vov, 0.85*gm_gds, label='gm/gds', color='green', linestyle='-')

plt.ylim(0, 50)
plt.xlabel("$V_{OV}$ (V)", fontsize=18)
plt.ylabel("FoM (a.u.)", fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
#plt.legend()
plt.show()
```
