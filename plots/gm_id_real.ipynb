```python
import numpy as np
import scipy.constants as sc
import matplotlib.pyplot as plt
from pygmid import Lookup as lk
```

```python
table_path = '/home/bmurmann/eda/designs/Book-on-gm-ID-design/starter_files_open_source_tools/ihp-sg13g2/simulation/'
nch = lk(table_path+'sg13_lv_nmos.mat')

vgs = np.linspace(0, 1.5, num=100)
gm_id_real = nch.lookup('GM_ID', VGS=vgs)
vt_real = nch.lookup('VT', VGS=vgs)
vov_real = vgs-vt_real


vov = np.linspace(0.001, 1, num=100)
gm_id = 2/vov

plt.plot(vov_real, gm_id_real,  label='$g_m/I_D$ real', color='blue', linestyle='-')
plt.plot(vov, gm_id, label='$g_m/I_D$ square law', color='blue', linestyle='--')
plt.xlim(-0.4, 1)
plt.ylim(0, 35)
plt.xlabel("$V_{OV}$ (V)", fontsize=18)
plt.ylabel("$g_m/I_D$ (S/A)", fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.legend(fontsize=14)
plt.show()
```
