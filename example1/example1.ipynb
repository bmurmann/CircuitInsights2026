```python
import numpy as np
import pandas as pd
from pygmid import Lookup as lk
```

#### Technology data

```python
table_path = '/home/bmurmann/eda/designs/Book-on-gm-ID-design/starter_files_open_source_tools/ihp-sg13g2/simulation/'
n = lk(table_path+'sg13_lv_nmos.mat')
```

#### Specifications

```python
gm = 10e-3
```

#### Design choices

```python
l = 0.13
gm_id = np.array([5, 15, 25])
```

#### Sizing and benchmarking

```python
id = gm/gm_id
jd = n.lookup('ID_W', GM_ID=gm_id, L=l)
w = id/jd

cgg_w = n.lookup('CGG_W', GM_ID=gm_id, L=l)
cgg = w*cgg_w
ft = gm/cgg/2/np.pi
```

```python
df = pd.DataFrame([gm_id, id, jd, w, cgg, ft], ['gm_id', 'id', 'jd', 'w', 'cgg', 'ft'], columns=['option1', 'option2', 'option3']); df
```
