```python
import numpy as np
import scipy.constants as sc
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('./xschem/simulation/example2.txt', sep='\\s+')
df
```

```python
f = df['frequency']
mag = df['vod_mag']
mag_db = 20*np.log10(mag)
mag_db_max = np.max(mag_db)

f_calc = [2e6, 65.9e6]
mag_db_calc = [12, mag_db_max-3]

plt.semilogx(f, mag_db, label='Simulation', color='blue', linestyle='-')
plt.semilogx(f_calc, mag_db_calc, label='Calculation', marker='+', markersize=15, markeredgewidth=2, linestyle='None', color='black')
plt.xlim(1e6, 1e10)
#plt.ylim(0, 35)
plt.xlabel("f (Hz)", fontsize=18)
plt.ylabel("Magnitude (dB)", fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.legend(fontsize=14, loc='lower left')
plt.show()
```
