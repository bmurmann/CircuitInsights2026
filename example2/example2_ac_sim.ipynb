```python
import numpy as np
import scipy.constants as sc
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('./xschem/simulations/example2.txt', sep='\\s+')
df
```

```python
f = df['frequency']
mag = df['vod_mag']
mag_db = 20*np.log10(mag)

plt.semilogx(f, mag_db, color='blue', linestyle='-')
plt.xlim(1e6, 1e10)
#plt.ylim(0, 35)
plt.xlabel("f (Hz)", fontsize=18)
plt.ylabel("Magnitude (dB)", fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.show()
```

```python
dc = np.max(mag_db)
bw = np.interp(dc-3, np.flip(mag_db), np.flip(f))
print(bw)
```
